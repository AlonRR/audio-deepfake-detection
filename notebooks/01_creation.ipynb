# 01 — Creation (voice cloning)

**This stage ran as Slurm batch jobs on the lab L4, not in this notebook** — the two
systems need different containers (XTTS needs PyTorch, the Keras baseline needs
TensorFlow), so they cannot share one kernel.

What actually ran:

| Step | Script | Job | Outcome |
|---|---|---|---|
| Dataset prep | `src/creation/xtts_finetune/prepare.py` | local | 34 clips / 4.27 min, 30 train + 4 held out |
| A2 fine-tune | `scripts/finetune_xtts.slurm` | 509 | 7 min; eval loss 3.825 → 3.485, best epoch 9 |
| A2 synthesis | `scripts/synth_xtts.slurm`, `synth_xtts_ft.slurm` | 511/513 | 12 zero-shot + 12 fine-tuned clips |
| A1 train + synth | `scripts/train_keras_tts.slurm` | 512 | 300 epochs, 1 h 07 m; 12 clips |
