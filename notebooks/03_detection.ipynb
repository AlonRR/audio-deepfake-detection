# 03 — Detection (anti-spoofing)

**Ran as Slurm jobs** (`scripts/base_cnn.slurm`, `extract_ssl.slurm`, `train_ssl.slurm`,
`score_*_eval.slurm`). Two containers: the frozen XLS-R frontend is PyTorch, the trainable
back-end is Keras, with features cached to disk between them.

Headline, on **unseen** eval attacks (A07–A19):

| System | dev EER | eval EER |
|---|---|---|
| SSL (XLS-R + Keras back-end) | 0.04% | **0.67%** |
| CNN-LFCC baseline | 12.99% | 18.55% |

Full numbers, the hyper-parameter sweep and the deliberate failed run:
`docs/results.md` §B and `reports/detection/`.
