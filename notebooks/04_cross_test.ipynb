# 04 — Cross-generator test (the contribution)

**Ran via `scripts/cross_test.slurm` (job 520).** Scores both detectors on our own
clips: 4 held-out real recordings vs 36 fakes from a 2025-era synthesizer.

| Detector | ASVspoof eval EER | On our clips | our fakes | our **real** |
|---|---|---|---|---|
| SSL | 0.67% | **29.17%** | 0.137 | **0.0009** |
| CNN-LFCC | 18.55% | **70.83%** | 0.086 | 0.057 |

Both collapse — and not as predicted. The SSL detector scores *genuine* speech at 0.0009
bona-fide, i.e. it is more confident the real recording is fake than the clones are. That
is **channel shift**, not generator novelty: it learned ASVspoof's VCTK recording
conditions rather than authenticity.

Limitation: n_real = 4, so directional not precise. Full write-up: `docs/results.md` §B.5.
